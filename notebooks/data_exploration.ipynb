```python
import pandas as pd
file_path= "../data/raw/Global Superstore.xls"
xls = pd.ExcelFile(file_path)
print(xls.sheet_names)
```

```python
orders = pd.read_excel(file_path, sheet_name="Orders")
returns = pd.read_excel(file_path, sheet_name="Returns")
people = pd.read_excel(file_path, sheet_name="People")

print("Orders:", orders.shape)
print("Returns:", returns.shape)
print("People:", people.shape)
```

```python
print("Orders Columns:")
print(orders.columns.tolist())
print("\nReturns Columns:")
print(returns.columns.tolist())
print("\nPeople Columns:")
print(people.columns.tolist())
```

```python
print("Orders Missing Values:")
print(orders.isnull().sum())

print("\nReturns Missing Values:")
print(returns.isnull().sum())

print("\nPeople Missing Values:")
print(people.isnull().sum())
```

```python
print("Orders Data Types:")
print(orders.dtypes)

print("\nReturns Data Types:")
print(returns.dtypes)

print("\nPeople Data Types:")
print(people.dtypes)
```

```python
print("Total rows:", len(orders))
print("Unique Row IDs:", orders["Row ID"].nunique())
print("Unique Order IDs:", orders["Order ID"].nunique())
print("Unique Customer IDs:", orders["Customer ID"].nunique())
print("Unique Product IDs:", orders["Product ID"].nunique())
```

```python
sample_order_id = orders["Order ID"].value_counts().index[0]

orders[orders["Order ID"] == sample_order_id][
    ["Order ID", "Order Date", "Customer ID", "Product ID", "Product Name", "Sales", "Quantity", "Profit"]
]
```
